```python
# Create two sets named hatfields and mccoys
hatfields = {"John", "William", "Sarah", "Rebecca", "Emma", "Andrew", "Isaac", "Mary", "Samuel", "Grace"}
mccoys = {"James", "Sarah", "Rebecca", "Emma", "Andrew", "Isaac", "Mary", "Thomas", "Samuel", "Grace"}

# Union of the two sets
union_set = hatfields.union(mccoys)
print("Union:", union_set)

# Intersection of the two sets
intersection_set = hatfields.intersection(mccoys)
print("Intersection:", intersection_set)

# Difference between hatfields and mccoys
difference_set = hatfields.difference(mccoys)
print("Difference (Hatfields - Mccoys):", difference_set)

# Symmetric difference between hatfields and mccoys
symmetric_difference_set = hatfields.symmetric_difference(mccoys)
print("Symmetric Difference:", symmetric_difference_set)
```

## Requirement 12

```python
def factorial(n):
    """
    Calculates the factorial of a number.
    
    Args:
    n (int): The number to calculate the factorial for.
    
    Returns:
    int: The factorial of the given number.
    """
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Test the factorial function with an example
result = factorial(5)
print("Factorial of 5 is:", result)
```

## Requirement 13

```python
def vegetables(*args, **kwargs):
    """
    Prints the numbers and names of vegetables passed as arguments.

    Args:
    *args: Variable-length argument list containing numbers of vegetables.
    **kwargs: Arbitrary keyword arguments containing names of vegetables.
    """
    for i, num_veggies in enumerate(args):
        print(f"Vegetable {i+1}: {num_veggies}")

    for veggie_name, veggie_num in kwargs.items():
        print(f"{veggie_name}: {veggie_num}")


# Example usage of the vegetables function
vegetables(3, 5, 7, potato=2, tomato=4, carrot=3)
```

```python
# Function to convert a string to an integer
def convert_to_int(s):
    try:
        result = int(s)
        print("Conversion successful. Integer value:", result)
    except ValueError:
        print("Error: Could not convert to integer. Invalid input!")

# Example usage of the convert_to_int function
convert_to_int("123")   # This should print the integer value 123
convert_to_int("abc")   # This should handle the ValueError
```

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Division by zero!")
    else:
        print("Division result:", result)
    finally:
        print("Division operation completed.")

# Example usage of the divide function
divide(10, 2)  # This should print the division result
print("--------------------")
divide(8, 0)   # This should handle the ZeroDivisionError
```

## Requirement 17

Iterator indirection refers to the process of indirectly accessing elements in a sequence through an iterator object rather than directly accessing the elements themselves. In Python, when iterating over a sequence like a list or a tuple, an iterator object is used to keep track of the current position in the sequence and provide access to the elements sequentially. Instead of directly accessing elements by index or position, the iterator object abstracts away the details of the underlying data structure and provides a consistent interface for accessing elements. This indirection allows for more flexible and efficient iteration, especially when dealing with large or complex data structures, and promotes code reusability and clarity by providing a unified approach to iterating over different types of sequences.
