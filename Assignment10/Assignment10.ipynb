## Requirement 9
In your own words, explain the difference between the keywords and and or
and the operators & and |. When are the keywords used and when are the
operators used?

The keywords and and or are used for boolean operations in Python. They are used to combine multiple conditions or expressions and evaluate them together.

* and: It returns True if both conditions or expressions on its left and right sides are True.
* or: It returns True if at least one of the conditions or expressions on its left and right sides is True.

The operators & and |, on the other hand, are bitwise operators in Python. They perform bitwise operations on integers and bitwise logical operations on boolean arrays.

* &: It performs a bitwise AND operation. When used with boolean arrays, it performs element-wise AND operation.
* |: It performs a bitwise OR operation. When used with boolean arrays, it performs element-wise OR operation.

Here's a summary of when to use each:

* Use 'and' and 'or' when you want to combine boolean conditions or expressions in Python code.
* Use '&' and '|' when you want to perform element-wise logical operations on boolean arrays in NumPy or when you want to perform bitwise operations on integers.

Here are examples illustrating the usage of both:

```python
import numpy as np

# Using 'and' keyword
x = 5
y = 10
c = -2
if x > 0 and y > 0:
    print("Both x and y are greater than 0: ", x, y)
if x > 0 or c > 0:
    print("At least one of x and y is greater than 0: ",x,c)

arr1 = np.array([True, False, True])
arr2 = np.array([False, True, True])

# Using '&' operator
result = arr1 & arr2
print("Result of element-wise AND operation:", result)

# Using '|' operator
result = arr1 | arr2
print("Result of element-wise OR operation:", result)
```

## Requirement 13
Produce a k-nearest neighbor plot using: 
* a 25 X 2 array 
* K = 3 


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

# Generating random data for demonstration
np.random.seed(0)
data = np.random.rand(25, 2)

# Creating a NearestNeighbors model with k=3
k = 3
nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(data)

# Finding the k-nearest neighbors for each point
distances, indices = nbrs.kneighbors(data)

# Plotting the k-nearest neighbor plot
plt.figure(figsize=(8, 6))
for i in range(len(data)):
    plt.scatter(data[i, 0], data[i, 1], color='blue', marker='o')  # Plotting the data point
    neighbors = data[indices[i][1:]]  # Excluding the first point (itself)
    plt.scatter(neighbors[:, 0], neighbors[:, 1], color='red', marker='x')  # Plotting the neighbors
plt.title('K-Nearest Neighbor Plot (K=3)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
