## Requirement 10
Demonstrate the use of wildcard matching. 

In Python, wildcard matching, also known as wildcard imports, allows you to import all objects (functions, classes, variables) from a module into the current namespace using the * character. However, wildcard imports are generally discouraged because they can lead to namespace pollution and make it unclear where certain objects are coming from.

```python
from math import *

# Now all functions and constants from the math module are available in the current namespace
```

In the above code, from math import * imports all objects from the math module into the current namespace. However, it's better to import specific objects explicitly to maintain code clarity and avoid potential conflicts. For example:

```python
from math import sin, cos, pi

# Now only sin, cos, and pi are available in the current namespace
```

## Requirement 17
Demonstrate the use %xmode magic function. 

The %xmode magic function in Jupyter Notebook allows you to control the amount of information displayed in the traceback when an exception occurs. It has three modes: Plain, Context, and Verbose. Let's demonstrate each mode:

```python
# Enable %xmode magic function with mode 'Plain'
%xmode Plain

# Define a function with a deliberate error
def divide_by_zero():
    return 1 / 0

# Call the function to trigger an exception
divide_by_zero()
```

```python
# Enable %xmode magic function with mode 'Context'
%xmode Context

# Call the function again to trigger an exception
divide_by_zero()
```

```python
# Enable %xmode magic function with mode 'Verbose'
%xmode Verbose

# Call the function again to trigger an exception
divide_by_zero()
```

## Requirement 20
Use at least 4 magic commands to time a function(s) that you have written. 

```python
def custom_function(n):
    """
    A custom function to calculate the sum of the first n natural numbers.
    """
    return sum(range(1, n+1))

# Using %time to time a single execution of the function
%time result = custom_function(1000000)

# Using %timeit to time multiple executions of the function and compute the average time
%timeit -n 100 custom_function(1000000)

# Using %%time to time the execution of a block of code containing function call
%%time
for _ in range(10):
    custom_function(1000000)

# Using %%timeit to time multiple executions of a block of code containing function call and compute the average time
%%timeit -n 10
for _ in range(10):
    custom_function(1000000)
```
